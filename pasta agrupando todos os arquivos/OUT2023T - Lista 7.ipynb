## Exercicio 1:


        (lista 5)
(1) Crie uma estrutura que pergunte ao usuário qual o dia da semana. Se o dia for igual a Domingo ou igual a sábado, imprima na tela "Hoje é dia de descanso", caso contrário imprima na tela "Você precisa trabalhar!"

```python
while True:
    inputDiaSemana = input("Que dia da semana é hoje? ('exit' para encerrar o programa.)").lower()
    if inputDiaSemana == 'exit':
        break
    diasDeDescanso = ['domingo','sábado']
    diasDeTrabalho = ['segunda','terça','quarta','quinta','sexta']
    try:
        if inputDiaSemana in diasDeDescanso:
            print("Hoje é dia de descanso.")
        elif inputDiaSemana in diasDeTrabalho:
            print("Você precisa trabalhar!")
        else:
            print("por favor insira um dia da semana válido e tente novamente")
    finally:    
        print("Tenha um bom dia.")
```

## Exercicio 2:


        (lista 5)
(5) Peça ao operador para inserir um numero. Imprima se o numero for um número primo ou não.

```python
while True:
    try:
        inputPrimos = input("Insira um número para verificar se ele é primo ou 'exit' para cancelar a operação: ")
        if inputPrimos == 'exit':
            break
        numero = int(inputPrimos)
        if numero <= 1:
            resultado = False
        elif numero <= 3:
            resultado = True
        elif numero % 2 == 0 or numero % 3 == 0:
            resultado = False
        else:
            i = 5
            resultado = True
            while i * i <= numero:
                if numero % i == 0 or numero % (i + 2) == 0:
                    resultado = False
                    break
                i += 6
        if resultado:
            print("Este é um número primo.")
        else:
            print("Este não é um número primo.")
    
    except ValueError:
        print("Por favor, insira um número válido.")
```

## Exercicio 3:


        (lista 5)
(9) Crie uma calculadora de parcelamento onde o código deve:
- receber o valor em reais 
- a quantidade de parcelas para realizar o calculo 
- o valor do salário do operador
- enviar para o operador os valores que foram calculados. 
O valor deve ser mostrado com todas as informações passadas pelo operador e deve conter o valor da parcela arredondado com 2 casas decimais.
- Se o valor das parcelas exceder 30% do salário informado pelo Operador, deve ser impresso na tela "Compra Negada"

```python
while True:
    try:
        valorEmReais = float(input("Informe o valor em reais.('exit' para cancelar a operação)"))
 
        qtdParcelas = int(input("Informe a quantidade de parcelas."))
        print(f"a quantidade de parcelas é: {qtdParcelas}")

        salarioOperador = float(input("Informe o valor de seu salário"))
        print(f"o salario do operador é: {salarioOperador}")

        valorParcelas = valorEmReais / qtdParcelas 
        print(f"sendo assim o valor das parcelas é {round(valorParcelas,2)}")

        pctSalarioOperador = salarioOperador * 0.3
        if valorParcelas > pctSalarioOperador:
            print("compra negada")
        elif valorParcelas < pctSalarioOperador:
            print("compra aprovada")
    except ValueError:
        print("por favor insira um valor valido.")
    finally:
        pass
    if valorParcelas or qtdParcelas or salarioOperador == 'exit':
            break
```

(11) Faça um programa que calcule a média ponderada de um aluno onde deve receber os valores da Prova 1 (P1= 0 ~ 100) e da Prova 2 (P2= 0 ~ 100), considerando que o peso da P1 é $0.4$ e o peso da P2 é $0.6$.
Se a nota for maior que 70 deve imprimir na tela "APROVADO", caso contrário "REPROVADO".

```python
try:
    nota_1= int(input("insira a nota da prova 1 (0 ~ 100)")) * .4
    nota_2= int(input("insira a nota de prova 2 (0 ~ 100)")) * .6
    total1 = nota_1 + nota_2
    print("APROVADO") if total1 > 70 else print ("REPROVADO")
except ValueError:
    print("não foi possivel concluir o cálculo das notas pois você digitou um valor incorreto.")
```
